```python
# Cell 1 — Env vars
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"

import pandas as pd
import numpy as np
```

```python
DATA_ROOT = os.path.expanduser(
    "~/Cancer-Multi-Omics-Benchmark/Main_Dataset/Classification_datasets/Pan-cancer/Original"
)

X_mrna  = pd.read_csv(f"{DATA_ROOT}/Pan-cancer_mRNA.csv",      index_col=0).T
X_mirna = pd.read_csv(f"{DATA_ROOT}/Pan-cancer_miRNA.csv",     index_col=0).T
X_methy = pd.read_csv(f"{DATA_ROOT}/Pan-cancer_Methy.csv",     index_col=0).T
X_cnv   = pd.read_csv(f"{DATA_ROOT}/Pan-cancer_CNV.csv",       index_col=0).T
labels  = pd.read_csv(f"{DATA_ROOT}/Pan-cancer_label_num.csv") # no index_col

# Reset all omics indices to integer positions to match labels
X_mrna.reset_index(drop=True, inplace=True)
X_mirna.reset_index(drop=True, inplace=True)
X_methy.reset_index(drop=True, inplace=True)
X_cnv.reset_index(drop=True, inplace=True)

print(f"mRNA  : {X_mrna.shape}")
print(f"miRNA : {X_mirna.shape}")
print(f"Methy : {X_methy.shape}")
print(f"CNV   : {X_cnv.shape}")
print(f"Labels: {labels.shape}")
print(f"\nAll indices aligned: {len(X_mrna) == len(labels)}")
```

```python
# Cell 3 — Concatenate and save
df_real = pd.concat([X_mrna, X_mirna, X_methy, X_cnv], axis=1)
df_real["cancer_type"] = labels["Label"].values

print(f"Final dataset shape: {df_real.shape}")
print(f"Cancer types: {df_real['cancer_type'].nunique()} unique")
print(f"Missing values: {df_real.isnull().sum().sum()}")

df_real.to_csv("cmob_real.csv", index=False)
print("Saved: cmob_real.csv")
```
